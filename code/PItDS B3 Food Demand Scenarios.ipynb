### Filter for only baseline scenario in Brazil

```python
animal_fd = food_demand.query('Scenario == "ba" and Country == "44 Brazil (BR)" and Food == ["13 Beef", "12 Pork", "11 Poultry", "15 Fish"]')
animal_fd.head(n=5)
```

```python
beef_fd = animal_fd = food_demand.query('Scenario == "ba" and Country == "44 Brazil (BR)" and Food == "13 Beef"')
beef_fd.head(n=5)
```

### Plot a single variable

```python
import matplotlib.pyplot as plt
import numpy as np
# Make sure figures appear inline in Ipython Notebook
%matplotlib inline
```

```python
plt.plot(beef_fd['Year'], beef_fd['Value (th tonnes)'])
plt.title('Beef Demand in Brazil')
plt.xlabel('Year')
plt.ylabel('Value in th tonnes')
plt.show()
```

### Plot multiple variables

```python
import seaborn as sns

x = animal_fd['Year']
y = animal_fd['Value (th tonnes)']
group = animal_fd['Food']

sns.scatterplot(x = x, y = y, hue = group)
```
