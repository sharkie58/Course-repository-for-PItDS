1. Import libraries

```python
# import science-kit learn and set up some shorthand
import sklearn as sk
import sklearn.datasets as skd
import sklearn.cluster as skc

# import matplotlib for plots
import matplotlib.pyplot as plt
```

2. Create a simple dataset

```python
# create a random test dataset with some “blobs” 
# to see if we can find them with a clustering algorithm
X, y = skd.make_blobs()
```

```python
# show this dataset
plt.scatter(X[:,0], X[:,1])
plt.show()
```

3. Use k-means to create 3 clusters

```python
# use the built-in KMeans implementationin scikit-learn 
# to create 3 clusters and store in a variable called model
model = skc.KMeans(n_clusters = 3)

# see the score of how well the algorithm clustered the data
model.fit(X).score(X)
```

```python
# color to points accordingly
plt.scatter(X[:,0], X[:,1], c = model.labels_)
```

In real data, you wouldn’t necessarily have a “true” or “correct” value for each point in the dataset, but since we createdartificial data with make_blobs(), we can actually look at the “real” values. These were stored in y.

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

4. Experiment with other data

```python
# Make different blobs
A, b = skd.make_blobs(centers = 9)
plt.scatter(A[:,0], A[:,1])
plt.show()
```

```python
blobs = skc.KMeans(n_clusters = 5)
blobs.fit(A).score(X)
```

```python
plt.scatter(A[:,0], A[:,1], c = blobs.labels_)
```

```python
plt.scatter(A[:,0], A[:,1], c = b)
```

```python
# Make moons
C, d = skd.make_moons()
plt.scatter(C[:,0], C[:,1])
plt.show()
```

```python
moons = skc.KMeans(n_clusters=2)
moons.fit(C)
```

```python
plt.scatter(C[:,0],C[:,1],c=moons.labels_)
```

```python
# Make circles
E, f = skd.make_circles()
plt.scatter(E[:,0], E[:,1])
plt.show()
```

```python
circles = skc.KMeans(n_clusters=2)
circles.fit(E)
```

```python
plt.scatter(E[:,0],E[:,1],c=circles.labels_)
```

5. Try DBSCAN

```python
db = skc.DBSCAN(eps=2, min_samples = 10)
db.fit(C)
plt.scatter(C[:,0], C[:,1],c=db.labels_)
```
